```python
#9 Leia uma temperatura e a unidade de origem usando match/case ("C" para Celsius, "F" para Fahrenheit ou "K" para Kelvin). Valide que temperaturas em Kelvin não sejam negativas. Calcule e exiba a temperatura convertida nas outras duas escalas. Use variáveis descritivas como temp_celsius, temp_fahrenheit e temp_kelvin. Use case _ para unidade inválida.
def converter_temperatura():
    try:
        entrada_temp = float(input("Digite a temperatura: "))
        unidade = input("Digite a unidade de origem (C, F, K): ").upper()
        match unidade:
            case "C":
                temp_celsius = entrada_temp
                temp_fahrenheit = (temp_celsius * 9/5) + 32
                temp_kelvin = temp_celsius + 273.15
            case "F":
                temp_fahrenheit = entrada_temp
                temp_celsius = (temp_fahrenheit - 32) * 5/9
                temp_kelvin = temp_celsius + 273.15
            case "K":
                temp_kelvin = entrada_temp
                if temp_kelvin < 0:
                    print("Erro: A temperatura em Kelvin não pode ser negativa.")
                    return
                temp_celsius = temp_kelvin - 273.15
                temp_fahrenheit = (temp_celsius * 9/5) + 32
            case _:
                print("Erro: Unidade de medida inválida. Use 'C', 'F' ou 'K'.")
                return
        print(f"\nResultados:")
        print(f"Celsius: {temp_celsius:.2f}°C")
        print(f"Fahrenheit: {temp_fahrenheit:.2f}°F")
        print(f"Kelvin: {temp_kelvin:.2f}K")

    except ValueError:
         print("Erro: Entrada inválida. Por favor, digite um número para a temperatura.")
converter_temperatura()
```

```python
#14 Leia as quatro notas bimestrais de um aluno (0 a 10 cada). Valide todas. Calcule a média aritmética. Use elif para o conceito: média ≥ 9 → A, ≥ 7 → B, ≥ 5 → C, ≥ 3 → D, abaixo → E. Com if aninhado, exiba o status: A e B → "Aprovado", C → "Recuperação", D e E → "Reprovado". Exiba a média com 2 casas decimais e o conceito.
notas = []
for i in range(1, 5):
    while True:
        nota = float(input(f"Digite a {i}ª nota (0 a 10): "))
        if 0 <= nota <= 10:
            notas.append(nota)
            break
        print("Nota inválida! Digite um valor entre 0 e 10.")
media = sum(notas) / 4
if media >= 9:
    conceito = "A"
elif media >= 7:
    conceito = "B"
elif media >= 5:
    conceito = "C"
elif media >= 3:
    conceito = "D"
else:
    conceito = "E"
if conceito == "A" or conceito == "B":
    status = "Aprovado"
else:
    if conceito == "C":
        status = "Recuperação"
    else:
        status = "Reprovado"
print('-'*40)
print(f"\nMédia: {media:.2f}")
print(f"Conceito: {conceito}")
print(f"Status: {status}")
```

```python
#29 Leia o tipo do atacante e o tipo do defensor usando match/case. Tipos: "fogo", "agua", "planta", "eletrico", "terra". Relações de vantagem: fogo vence planta, água vence fogo, planta vence água, elétrico vence água, terra vence elétrico. Com if aninhado, determine se é vantagem (dano x2), desvantagem (dano x0.5) ou neutro (dano x1). Leia o poder base do ataque, calcule o dano final e exiba o resultado.
def calcular_dano():
    print("Tipos disponíveis: fogo, agua, planta, eletrico, terra")
    atacante = input("Tipo do atacante: ").lower().strip()
    defensor = input("Tipo do defensor: ").lower().strip()
    try:
        poder_base = float(input("Poder base do ataque: "))
    except ValueError:
        print("Poder base inválido.")
        return
    tipos_validos = ["fogo", "agua", "planta", "eletrico", "terra"]
    if atacante not in tipos_validos or defensor not in tipos_validos:
        print("Tipo inválido. Use apenas: fogo, agua, planta, eletrico, terra.")
        return
    multiplicador = 1.0
    mensagem = "Dano neutro!"
    match atacante:
        case "fogo":
            if defensor == "planta":
                multiplicador = 2.0
                mensagem = "Vantagem! (Super efetivo)"
            elif defensor == "agua" or defensor == "fogo":
                multiplicador = 0.5
                mensagem = "Desvantagem! (Pouco efetivo)"
        case "agua":
            if defensor == "fogo" or defensor == "terra":
                multiplicador = 2.0
                mensagem = "Vantagem! (Super efetivo)"
            elif defensor == "planta" or defensor == "agua":
                multiplicador = 0.5
                mensagem = "Desvantagem! (Pouco efetivo)"
        case "planta":
            if defensor == "agua" or defensor == "terra":
                multiplicador = 2.0
                mensagem = "Vantagem! (Super efetivo)"
            elif defensor == "fogo" or defensor == "planta":
                multiplicador = 0.5
                mensagem = "Desvantagem! (Pouco efetivo)"
        case "eletrico":
            if defensor == "agua":
                multiplicador = 2.0
                mensagem = "Vantagem! (Super efetivo)"
            elif defensor == "terra" or defensor == "eletrico" or defensor == "planta":
                multiplicador = 0.5
                mensagem = "Desvantagem! (Pouco efetivo)"
        case "terra":
            if defensor == "eletrico" or defensor == "fogo":
                multiplicador = 2.0
                mensagem = "Vantagem! (Super efetivo)"
            elif defensor == "planta":
                multiplicador = 0.5
                mensagem = "Desvantagem! (Pouco efetivo)"
    dano_final = poder_base * multiplicador
    print("-" * 30)
    print(f"Atacante ({atacante}) vs Defensor ({defensor})")
    print(f"Resultado: {mensagem}")
    print(f"Multiplicador: x{multiplicador}")
    print(f"Dano Final: {dano_final}")
    print("-" * 30)
calcular_dano()
```

```python
#30 Leia o tipo de acomodação usando match/case ("standard", "luxo", "suite"), o número de diárias (validar como positivo) e se o hóspede tem cartão fidelidade (s/n). Defina a diária base de cada tipo como variável. Use elif aninhado para desconto por diárias: 1-3 → 0%, 4-7 → 10%, acima de 7 → 20%. Com if aninhado adicional, aplique mais 5% para cartão fidelidade. Valide que o total não seja inferior a R$100. Use variáveis descritivas (diaria_base, desconto_qtd, desconto_fidelidade, total_final) e exiba o resumo completo com f-string.
tipo_acomodacao = input("Digite o tipo de acomodação (standard, luxo, suite): ").lower().strip()
diarias = int(input("Digite o número de diárias: "))
if diarias <= 0:
    print("Erro: O número de diárias deve ser positivo.")
else:
    fidelidade = input("Hóspede tem cartão fidelidade? (s/n): ").lower().strip()
    match tipo_acomodacao:
        case "standard":
            diaria_base = 150.0
        case "luxo":
            diaria_base = 300.0
        case "suite":
            diaria_base = 500.0
        case _:
            diaria_base = 0.0
            print("Tipo de acomodação inválido.")
    if diaria_base > 0:
        if 1 <= diarias <= 3:
            desconto_qtd = 0.0
        elif 4 <= diarias <= 7:
            desconto_qtd = 0.10
        else:
            desconto_qtd = 0.20
        subtotal = diarias * diaria_base
        valor_desconto_qtd = subtotal * desconto_qtd
        total_após_desconto = subtotal - valor_desconto_qtd
        desconto_fidelidade = 0.0
        if fidelidade == 's':
            desconto_fidelidade = total_após_desconto * 0.05
        total_final = total_após_desconto - desconto_fidelidade
        if total_final < 100:
            total_final = 100.0
            aviso = "(Valor mínimo de R$100 aplicado)"
        else:
            aviso = ""
        print("-" * 30)
        print("RESUMO DA RESERVA")
        print("-" * 30)
        print(f"Acomodação: {tipo_acomodacao.capitalize()}")
        print(f"Diárias: {diarias}")
        print(f"Diária Base: R$ {diaria_base:.2f}")
        print(f"Subtotal: R$ {subtotal:.2f}")
        print(f"Desconto Qtd ({desconto_qtd*100:.0f}%): -R$ {valor_desconto_qtd:.2f}")
        if fidelidade == 's':
            print(f"Desconto Fidelidade (5%): -R$ {desconto_fidelidade:.2f}")
        print(f"Total Final: R$ {total_final:.2f} {aviso}")
        print("-" * 30)
```
